# Guided Project: Exploring Hacker News Posts

```python
import pandas as pd
```

```python
from csv import reader
opened_file = open('hacker_news.csv')
read_file = reader(opened_file)
hn = list(read_file)
hn[:5]
```

```python
headers = hn[0]
headers
```

```python
hn = hn[1:]
```

```python
ask_posts = []
show_posts = []
other_posts = []

for row in hn:
    title = row[1]
    if title.lower().startswith('ask hn'):
        ask_posts.append(row)
    if title.lower().startswith('show hn'):
        show_posts.append(row)
    else:
        other_posts.append(row)
    
    
```

```python
len(ask_posts), len(other_posts), len(show_posts)
```

```python
show_posts
```

```python
total_ask_comment = 0
count = 0
for row in ask_posts:
    num_comments = int(row[4])
    count += 1
    total_ask_comment += num_comments
avg_ask_comments = total_ask_comment / count

total_show_comment = 0
count = 0
for row in show_posts:
    num_comments = int(row[4])
    count += 1
    total_show_comment += num_comments
avg_show_comments = total_show_comment / count

print('Average of ask comment point:{}\nAverage of show comment point:{}'.format(avg_ask_comments, avg_show_comments))
    
```

As it can be seen above, ask posts receives more comments than show posts receives on average.

## Calculate when comments created using datetime

```python
import datetime as dt
result_list = []
for row in ask_posts:
    create_time = row[6]
    comments = int(row[4])
    result_list.append([create_time, comments])
```

```python
counts_by_hour = {}
comments_by_hour = {}
for row in result_list:
    date = dt.datetime.strptime(row[0],  "%m/%d/%Y %H:%M")
    hour = date.strftime("%H")
    if hour not in counts_by_hour:
        counts_by_hour[hour] = 1
        comments_by_hour[hour] = row[1]
    else:
        counts_by_hour[hour] += 1
        comments_by_hour[hour] += row[1]
        
    
    
```

```python
comments_by_hour
```

```python
total_comment = sum(comments_by_hour.values())
total_comment
```

```python
avg_by_hour = []
for hour in comments_by_hour:
    avg_by_hour.append([hour, comments_by_hour[hour] / total_comment * 100])
    
```

```python
swap_avg_by_hour = []
for row in avg_by_hour:
    swap_avg_by_hour.append([row[1], row[0]])
```

```python
swap_avg_by_hour
```

```python
sorted_swap = sorted(swap_avg_by_hour, reverse=True)
```

```python
for row in sorted_swap[:5]:
    hour = dt.datetime.strptime(row[1], "%H")
    print("{}: {:.2f} average comments per post".format(hour.strftime("%H:%M"), row[0] ))
```
