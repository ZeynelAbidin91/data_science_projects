```python
dete_resignations = dete_survey_updated[dete_survey_updated['separationtype'].str.contains(r'^Resignation')]
```

```python
dete_resignations.head()
```

```python
tafe_resignation = tafe_survey_updated[tafe_survey_updated['separationtype'] == 'Resignation']
```

```python
tafe_resignation.head()
```

```python
dete_resignations['cease_date'].value_counts()
```

```python
dete_resignations['cease_date'] = dete_resignations['cease_date'].str.split('/').str[-1].astype(float)
```

```python
boxplot = dete_resignations.boxplot(column=['cease_date', 'dete_start_date'])
```

```python
dete_resignations['institute_service'] = dete_resignations['cease_date'] - dete_resignations['dete_start_date']
```

```python
tafe_resignation['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
tafe_resignation['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
def update_vals(value):
    if pd.isnull(value):
        return np.nan
    elif value == '-':
        return False
    else:
        return True
```

```python
tafe_resignation['dissatisfied'] = tafe_resignation[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].applymap(update_vals).any(1, skipna=False)
tafe_resignations_up = tafe_resignation.copy()
```

```python
tafe_resignations_up['dissatisfied'].value_counts(dropna=False)
```

```python
dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction',
       'dissatisfaction_with_the_department', 'physical_work_environment',
       'lack_of_recognition', 'lack_of_job_security', 'work_location',
       'employment_conditions', 'work_life_balance',
       'workload']].any(1, skipna=False)
dete_resignations_up = dete_resignations.copy()
dete_resignations_up['dissatisfied'].value_counts(dropna=False)
```

```python
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index=True)
```

```python
combined.notnull().sum().sort_values()
```

```python
combined_updated = combined.dropna(axis=1,thresh=500).copy()
```

```python
combined_updated.notnull().sum().sort_values()
```

```python
combined_updated['institute_service_up'] = combined_updated['institute_service'].astype(str).str.extract(r'(\d+)')
```

```python
combined_updated['institute_service_up'].value_counts()
```

```python
combined_updated['institute_service_up'] = combined_updated['institute_service_up'].astype('float')
```

```python
def transform_service(val):
    if val >= 11:
        return "Veteran"
    elif 7 <= val < 11:
        return "Established"
    elif 3 <= val < 7:
        return "Experienced"
    elif pd.isnull(val):
        return np.nan
    else:
        return "New"
```

```python
combined_updated['service_cat'] = combined_updated['institute_service_up'].apply(transform_service)
```

```python
combined_updated['service_cat'].value_counts()
```

```python
combined_updated['dissatisfied'].value_counts(dropna=False)
```

```python
combined_updated['dissatisfied'] = combined_updated['dissatisfied'].fillna(False)
```

```python
dis_pct = combined_updated.pivot_table(index='service_cat', values='dissatisfied')
```

```python
dis_pct.head()
```

```python
%matplotlib inline
dis_pct.plot(kind='bar', rot=30)
```
